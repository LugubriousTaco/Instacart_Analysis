# If-Statements with User-Defined Functions

```python
#import libraries
import pandas as pd
import numpy as np
import os
```

```python
#import dataset
path = r'C:\Users\mmreg\OneDrive\Desktop\Data Analytics Course Work\Data Immersion\Tasks\04-2022 Instacart Basket Analysis'
df_ords_prods_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```

```python
#create smaller subset
df = df_ords_prods_merged[:1000000]
#this gathers first million rows of dataset
```

```python
#define function
def price_label(row):
    if row['prices']<=5:
        return 'Low-Range Product'
    elif (row['prices']>5) and (row['prices']<=15):
        return 'Mid-Range Product'
    elif row['prices']>15:
        return 'High-Range Product'
    else: return 'Not Enough Data'
```

```python
#apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#count price point flags
df['price_range'].value_counts(dropna=False)
```

```python
df['prices'].max()
```

# If-Statements with For-Loops

```python
#Determine frequency of days
df_ords_prods_merged['order_dow'].value_counts(dropna = False)
```

```python
#Run for-loop statement
result = []

for value in df_ords_prods_merged['order_dow']:
    if value == 0:
        result.append('busiest day')
    elif value == 4:
        result.append('least busy')
    else:
        result.append('regularly busy')
```

```python
#Create new column based on result
df_ords_prods_merged['busiest_day'] = result
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

## Question 4

### When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
#determine volume of orders by hour
df_ords_prods_merged.columns
```

```python
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
#write function to create variable
tod_result = []

for value in df_ords_prods_merged['order_hour_of_day']:
    if value > 8 & value < 17:
        tod_result.append('Most Orders')
    elif value <= 6:
        tod_result.append('Fewest Orders')
    elif value == 23:
        tod_result.append('Fewest Orders')
    else:
        tod_result.append('Average Orders')
```

```python
tod_result
```

```python
#create new column for variable
df_ords_prods_merged['busiest_period_of_day'] = tod_result
df_ords_prods_merged.head()
```
